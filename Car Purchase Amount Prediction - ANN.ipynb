```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Data Cleaning
X = car_df.drop(['Customer Name', 'Customer e-mail', 'Country', 'Car Purchase Amount'], axis =1)
# Dropping the columns which are not relevant and will not impact the outcome, Car Purchase Amount.
X
```

```python
Y = car_df['Car Purchase Amount']
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
# Normalize the data before training the data between 0 to 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled #data is normalized now
```

```python
X_scaled.shape
```

```python
scaler.data_max_
```

```python
scaler.data_min_
```

```python
Y.shape
```

```python
# Y needs to be reshaped before calling fit transform

Y = Y.values.reshape(-1,1)
#Reshape your data using array.reshape(-1, 1) if your data has a single feature
Y_scaled = scaler.fit_transform(Y)
Y_scaled
```

```python
#Training the Model
# Default value of test set is 25%
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y_scaled, test_size = 0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Faced some issue updating tensorflow with pip. Had to use 'pip install --ignore-installed tensorflow'
import tensorflow as tf
from tensorflow import keras
```

```python
# Perform Training
# Build ANN in sequential form. Fully connect layer ANN
# An input layer, one of more hidden layer and an Output layer

import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense


model = Sequential()
# No of neurons in first layer = 25, no of input is 5, Activation func is relu
model.add(Dense(25, input_dim = 5, activation = 'relu')) #Specify Whats my input, how many neuron in first layer, what my activation function will be.
model.add(Dense(25, activation = 'relu')) #Add additional hidden layer
model.add(Dense(1, activation = 'linear'))# output layer is 1, and activation func is linear


```

```python
model.summary()
```

```python
# Explanation on Parameters
# 25 Neurons and 150 param, 25 * 5 is 125 weights + 25 bias leads to 150 param
# Next layer, 25 * 25 neurons will generate 625 weights + 25 bias leads to 650
# The output layer has 25 neurons fully connected to output, 25 weights + 1 bias leads to 26
```

```python
#Train the Model
model.compile(optimizer  = 'adam', loss = 'mean_squared_error')
```

```python
epochs_hist = model.fit(X_train, Y_train, epochs = 100, batch_size = 25, verbose = 1, validation_split = 0.2)
```

```python
#Evaluate the model
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss progress during Training')
plt.ylabel('Training and Validation loss')
plt.xlabel('Epoch number')
plt.legend(['Training Loss', 'Validation loss'])
```

```python
# If you reduce the number of neurons, it will take more time to learn. The loss between training and validation will not dimish sooner.
# Try with different number of neurons and different epochs and see the results.
```

```python
# Evaluate the model with an input
#Gender, Age, Annual Salary, Credit Card Debt, Net worth
X_test = np.array([[1, 40, 45000, 20000, 150000]])
Y_predict = model.predict(X_test)

print('Predicted Car Purchase Amount', Y_predict)
```
